```python
import sqlite3
```

```python
conn = sqlite3.connect("music.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS songs (
    song_id INTEGER PRIMARY KEY,
    song_name TEXT,
    artist TEXT,
    genre TEXT,
    duration INTEGER
)
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    user_id INTEGER PRIMARY KEY,
    name TEXT,
    email TEXT
)
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS playlists (
    playlist_id INTEGER PRIMARY KEY,
    user_id INTEGER,
    song_id INTEGER
)
""")
conn.commit()
```

```python
songs = [
    ('Believer', 'Imagine Dragons', 'Rock', 3),
    ('Shape of You', 'Ed Sheeran', 'Pop', 4),
    ('Blinding Lights', 'The Weeknd', 'Pop', 3)
]
cursor.executemany("""
INSERT INTO songs (song_name, artist, genre, duration)
VALUES (?, ?, ?, ?)
""", songs)
conn.commit()
```

```python
users = [
    ('Bhargav', 'bhargav@gmail.com'),
    ('Rahul', 'rahul@gmail.com')
]
cursor.executemany("""
INSERT INTO users (name, email)
VALUES (?, ?)
""", users)
conn.commit()
```

```python
playlists = [
    (1, 1),
    (2, 2)
]
cursor.executemany("""
INSERT INTO playlists (user_id, song_id)
VALUES (?, ?)
""", playlists)
conn.commit()
```

```python
cursor.execute("""
SELECT users.name,
songs.song_name,
songs.artist
FROM playlists
JOIN users
ON playlists.user_id = users.user_id
JOIN songs
ON playlists.song_id = songs.song_id
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute("""
SELECT genre, COUNT(*)
FROM songs
GROUP BY genre
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
conn.close()
```
